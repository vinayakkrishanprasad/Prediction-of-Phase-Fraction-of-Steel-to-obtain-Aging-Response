```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import pandas as pd
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y,random_state=1, test_size=0.2)
```

```python
sc_X = StandardScaler()
X_trainscaled=sc_X.fit_transform(X_train)
X_testscaled=sc_X.transform(X_test)
```

```python
reg = MLPRegressor(hidden_layer_sizes=(64,16,8),alpha=0.0001,solver='adam',activation='relu',random_state=1, max_iter=1300,verbose=1, early_stopping=False).fit(X_trainscaled, y_train)
```

```python
y_pred=reg.predict(X_trainscaled)
print("The Score with ", (r2_score(y_pred, y_train)))
```

```python
y_pred=reg.predict(X_testscaled)
print("The Score with ", (r2_score(y_pred, y_test)))
```

# to find the best parameters we do a grid search


```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'hidden_layer_sizes': [(64,16,8), (100,50,20), (32,16,8)],
    'max_iter': [500, 1000,1300,1500],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant','adaptive'],
}
```
